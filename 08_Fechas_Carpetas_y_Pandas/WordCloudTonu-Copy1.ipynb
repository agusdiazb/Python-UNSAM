#### - Importamos primero las librerías que vamos a utilizar

```python
# Import necessary libraries
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt

%matplotlib inline
```

#### - Creamos una función para realizar el wordcloud

```python
#Create a function
def make_wordcloud (path_file, WC_name):
    '''
    Returns a wordcloud given a path file and the wordcloud name
    '''
    #Open and read File
    f = open(path_file)
    file_exp = f.read()
    
    #Download spanish stopwords:
    stops = set(stopwords.words('spanish'))

    #Update specific words to avoid
    stops.update(['creo', 'materia', 'cátedra', 'si', 'gustaría', 'hace','ma','mas',
              'bien', 'etc', 'persona', 'cada','expectativa', 'manera', 'gustaria',
              'poder', 'tener','tema', 'clase'])

    #Create WordCloud
    wordcloud = WordCloud (
        stopwords=stops,
        width = 1700,
        height = 900,
        max_font_size=200,
        max_words=500,
        background_color="white"
                          ).generate(file_exp)

    # Display the generated image:
    plt.rcParams["figure.figsize"] = (30,10)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(WC_name, fontdict = {'fontweight' : 10, 'style':'italic'}, fontsize = 20)
    plt.show()
    
    # Save the image in the img folder:
    wordcloud.to_file('expectativas.png')
```

```python
#Display WordCloud GENERAL
make_wordcloud('expectativas_cursada.txt', 'Expectativas de cursada')
```
